```python
Oe=OrdinalEncoder()
closInOe=["Gender",'Occupation','Smoking','Family History of Anxiety','Dizziness','Medication','Recent Major Life Event']
data[closInOe]=Oe.fit_transform(data[closInOe])
data
```

```python
X=data.drop('Anxiety Level (1-10)',axis=1)
y=data['Anxiety Level (1-10)']
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=42)
```

```python
Lr=LinearRegression()
```

```python
Lr.fit(X_train,y_train)
```

```python
y_pred=Lr.predict(X_train)
```

```python
print('mean absolute error:', mean_absolute_error(y_train, y_pred))
print('Mean squared error:',mean_squared_error(y_train,y_pred))
print('r2_score:',r2_score(y_train,y_pred))
```

```python
plt.scatter(y_train, y_pred, alpha=0.6)
plt.xlabel('Actual Anxiety Level')
plt.ylabel('Predicted Anxiety Level')
plt.title('Actual vs Predicted Anxiety Level')
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'r--')  # line of perfect prediction
plt.grid(True)
plt.show()
```

```python
residuals = y_train - y_pred
plt.scatter(y_pred, residuals, alpha=0.5)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Predicted')
plt.ylabel('Residuals')
plt.title('Residuals vs Predicted')
plt.grid(True)
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=400, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_train)
```

```python
print('r2_score:',r2_score(y_train,y_pred_rf))
```

```python
importances = rf.feature_importances_
features = X_train.columns
plt.barh(features, importances)
plt.title("Feature Importance")
plt.show()
```

```python
print("mean squared error:",mean_squared_error(y_train,y_pred_rf))
print("mean_absolute_error:",mean_absolute_error(y_train,y_pred_rf))
```
